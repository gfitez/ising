```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import ellipk
from scipy.integrate import quad
```

```python
#Numerically calculate c_V from energy
def Kappa(T):
  return 2*np.sinh(2/T)/(np.cosh(2/T))**2

def K1(T):
  return ellipk(Kappa(T))

def E(T):
  return -2*np.tanh(1/T)-(((np.sinh(2/T))**2-1)/(np.sinh(2/T)*np.cosh(2/T)))*(2*K1(T)/np.pi-1)

temps=np.linspace(1,4,1000) #replace with appropriate temps, this was just to make sure code is working
lefttemps=np.linspace(1,2.269,1000) #again this was just to test, will be whatever the lefttemps array is, this is for mag and chi
#differentiate wrt T
def cV(temps):
  energies=E(temps)
  cv_array=np.gradient(energies)
  return cv_array
```

```python
#calculate magnetization
def M(T):
  return np.abs(1-(np.sinh(2/T)**-4))**(1/8)
```

```python
plt.plot(temps,cV(temps))
plt.axvline(2.269,ls='--',c='k')
plt.show()

plt.plot(lefttemps,M(lefttemps),c='r')
plt.vlines(2.269,0,1,ls='--',color='k')
plt.hlines(0,2.269,4,color='r')
plt.show()
```
